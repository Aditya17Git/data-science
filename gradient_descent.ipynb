```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris=sns.load_dataset('iris')
iris
```

```python
x=iris["petal_length"]
y=iris["petal_width"]
```

```python
x=np.array(x)
x
```

```python
y=np.array(y)
y
```

```python
iterations=1000
learning_rate=0.05
m_curr=0
b_curr=0
n=len(x)
```

```python
def gradient_descent(x,y):
    n=len(x)
    learning_rate
    iterations
    m_curr
    b_curr
```

```python
 for i in range(iterations):
    yp=m_curr*x+b_curr
    cost=(1/n)*sum([val**2 for val in (y-yp)])
    md=-(2/n)*sum(x*(y-yp))
    bd=-(2/n)*sum((y-yp))
    m_curr=m_curr-learning_rate*md
    b_curr=b_curr-learning_rate*bd
    print("m  {},   b  {},   cost  {},   iterations  {}".format(m_curr,b_curr,cost,i))
```

```python
gradient_descent(x,y)
```
