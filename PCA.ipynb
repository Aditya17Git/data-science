# Dimensionality Reduction: PCA

```python
import pandas as pd
```

```python
df=pd.read_csv("Heart.csv")
df
```

```python
df['ChestPain']=df['ChestPain'].astype('category')
df['ChestPain']=df['ChestPain'].cat.codes

df['Thal']=df['Thal'].astype('category')
df['Thal']=df['Thal'].cat.codes

df['AHD']=df['AHD'].astype('category')
df['AHD']=df['AHD'].cat.codes
df
```

```python
df.isna().sum()
```

```python
df=df.dropna()
df
```

```python
x=df.drop(columns=['Unnamed: 0','AHD'])
x
```

```python
y=df['AHD']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
x_scaled[:5]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=.30,random_state=1)
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_train)
y_pred[:5]
```

```python
y_train[:5]
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=3)
x_pca=pca.fit_transform(x_scaled)
x_pca.shape
```

```python
x_pca[:5]
```

```python
x_train_pca,x_test_pca,y_train,y_test=train_test_split(x_pca,y,test_size=0.3,random_state=1)
len(x_train_pca)
```

```python
model=LogisticRegression()
model.fit(x_train_pca,y_train)
model.score(x_train_pca,y_train)
```

```python
pca=PCA(0.95)
x_pca1=pca.fit_transform(x_scaled)
x_pca1.shape
```

```python
x_train_pca1,x_test_pca1,y_train,y_test=train_test_split(x_pca1,y,test_size=0.3,random_state=1)
len(x_train_pca1)
```

```python
model=LogisticRegression()
model.fit(x_train_pca1,y_train)
model.score(x_train_pca1,y_train)
```
